```python
#Declaracion de variables 

inicio=int(input("Inicio:  "))
incremento=int(input("Incremento:  "))
Fin=int(input("Fin:  "))

#El incremento se pone en float para que pueda aumentar de decimal en decimal

for i in range (inicio,Fin,incremento):
  print(i)
```

```python
#ciclo for 5 

tabla= int(input("Cual tabla desea:  "))

for cont in range(11):
  res=tabla*cont
  print(tabla," x " ,cont, "= " ,res)
```

```python
#ciclo for 6 - CICLOS SECUANCIALES

#Ciclo para la primera tabla 

tabla= int(input("Cual tabla desea:  "))

for cont in range(11):
  res=tabla*cont
  print(tabla," x " ,cont, "= " ,res)
```

```python
#ciclo for 7- CICLOS ANIDADOS

#Variables:

#conTabla= es el contador del 0 al 10 donde se repite en los ciclos internos 
#tabla=es la variable mas externa 
#contMdor= contador de repeticion 

tabla= int(input("Cuantas tablas desea:  "))

for conTabla in range(1,tabla+1):
  print("\nT A B L A    D E L   ",conTabla, "\n")
  for contMdor in range (1,11):
    res=conTabla*contMdor
    print(conTabla," x " ,contMdor, "= " ,res)
```

```python
#ciclo for 8- CICLOS ANIDADOS con WHILE

#Variables:

#conTabla= es el contador del 0 al 10 donde se repite en los ciclos internos 
#tabla=es la variable mas externa 
#contMdor= contador de repeticion 

x=1

tabla= int(input("Cual tabla desea:  "))

while x<=10:
  res=tabla*x
  print(tabla," x " ,x, "= " ,res)
```
