```python
import pandas as pd
```

Para crear nuestro primer Data Frame, debo hacer uso de los diccionarios de python, dentro de los diccionarios vamos a crear las listas que contendran los elementos de mi tabla

```python
notaSemillero=pd.DataFrame({"Nombres":['Pedro','darden','leidy','Luis','martha',
                                       'andrea','carlos','felipe','andres','Julian'],
                            
                            "Semestre":[10,9,7,8,10,9,10,8,9,8]})
print(notaSemillero)
```

Podemos incluir columnas adicionales a nuestro dataframe haciendo uso del metedo operador 

```python
notaSemillero["Nota1"]=3
notaSemillero
```

Añadir diferentes datos para cada fila

```python
notaSemillero['Nota2']=[3.1,4.2,3.3,4.5,4,4,3.8,5,4,3.1]
notaSemillero
```

Pandas nos provee de varios metodos para agregar columnas a nuestra tabla

insert()
assign()
loc()


```python
notaSemillero.insert(4,'Ciudad',['Manizales','pereira','Armenia','Cordoba','Cali', 'Popayan','Manizales','Pereira','Manizales','Popayan'])
notaSemillero
```
